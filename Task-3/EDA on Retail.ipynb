### Data Visualization

```python
plt.figure(figsize = (12,6))
plt.bar("Sub-Category", "Category", data = df, color = "orange")
plt.title("Category vs Sub-Category")
plt.xlabel("Sub-Category")
plt.ylabel("Category")
plt.xticks(rotation = 90)
plt.show()
```

From the above bar chart we can see the distribution of data according to categories & sub-categories

```python
df.hist(bins = 50, figsize = (20,15), color = "green")
plt.show()
```

Above histograms depict the amount of data points in each of the columns.

```python
profit_plot = (ggplot(df, aes(x = 'Sub-Category', y = 'Profit', fill = 'Sub-Category')) + geom_col() + coord_flip() + scale_fill_brewer(type ="div", palette = "Spectral") + theme_classic() + ggtitle("Bar Chart for Sub-Category & Profit"))
display(profit_plot)
```

By looking at the above horizontal bar graph we can see that the "Binders" sub-category has the highest amount of profit but also has suffered the highest amount of loss amongst all other sub-categories that are mentioned. We can also see that the "Tables" as well as the "Bookcases" Sub-Category has also suffered a huge loss than making much of a profit. Copiers sub-category has made the highest profit without any losses. While, in the case of Machines, we can see that it's neither making any profit as well.

```python
ggplot(df, aes(x ="Ship Mode", fill ="Category")) + geom_bar(stat = "count") + scale_colour_brewer(palette ="Greens")
```

From the above stacked bar chart we can able to visualize the Category-wise count of data w.r.t different Ship Modes.

```python
figsize = (15,15)
sns.pairplot(df, hue = "Sub-Category")
plt.show
```

Pairplot really helps to find the correlation among different data among different Sub-Categories

```python
flip_xlabels = theme(axis_text_x = element_text(angle = 90, hjust = 1), figure_size = (10,5), axis_ticks_length_major = 10, axis_ticks_length_minor =5)
(ggplot(df, aes(x = "Sub-Category", fill = "Sales")) + geom_bar() + facet_wrap(['Segment']) + flip_xlabels + theme(axis_text_x = element_text(size = 12)) + ggtitle("Sales from Every Segment of United States"))
```

From the above graph it is clear that the Consumer segment has the highest number Sub-Category wise sales & the Home office segment has the least.

```python
flip_xlabels = theme(axis_text_x = element_text(angle = 90, hjust = 10), figure_size = (10,10), axis_ticks_length_major = 50, axis_ticks_length_minor =50)
(ggplot(df, aes(x = "Category", fill = "Sales")) + geom_bar() + facet_wrap(['Region']) + flip_xlabels + theme(axis_text_x = element_text(size = 12)) + ggtitle("Sales of each region  of United States"))
```

```python
plt.figure(figsize =(10,4))
sns.lineplot("Discount", "Profit", data = df, color = "b", label = "Discount")
plt.legend()
plt.show()
```

From the line plot above it is clear that the increase in Discount is getting the profits go down.
